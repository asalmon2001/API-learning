## Getting historical data

Lets see if I can get historical data and plot it.

```python
# First, get the data. lets start with yesterday
params = { 'key' : str(api_key) , 'q' : 'Calgary' , 'dt' : "2023-09-13"}
response_yesterday = requests.get("http://api.weatherapi.com/v1/history.json", params=params)
print(response_yesterday.status_code)
```

```python
# Lets see how it is formatted
str_yesterday = json.dumps(obj=response_yesterday.json(), indent=2)
print(str_yesterday)
```

```python
# Get the dict
yesterday = response_yesterday.json()
```

```python
# First, lets get the forecast
forecast = yesterday['forecast']['forecastday']
# Now, the hour by hour data
hourly = forecast[0]['hour']
# Now, lets get the time and temperature in Celsius
time_temp = [[datetime.fromisoformat(current['time']).time() for current in hourly], [current['temp_c'] for current in hourly]]
```

```python
df = pd.DataFrame(time_temp).transpose()
print(df)
```

So, I just learned that you can just hand pandas the hourly data and it handles it perfectly. I forgot how powerful pandas was...
